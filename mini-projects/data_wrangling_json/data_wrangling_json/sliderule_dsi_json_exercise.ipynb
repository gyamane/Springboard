## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
#Load 'data/world_bank_projects' as a string of data
project_data = json.load((open('data/world_bank_projects.json')))

#Use normalization to create data of major themes from nested element in the above string of data
grouped_themes_df = json_normalize(project_data, 'mjtheme_namecode')
grouped_themes_df
```

```python
grouped_themes_df['name'].value_counts()
```

This made it so I could reference it in my call to **json_normalize()** to obtain a dataframe from the nested *'mjtheme_namecode'* data. From this resulting dataframe, I could follow a similar approach to question 1, where I referenced the 'name' column to get the names of the project themes and grouped by their frequency to easily see which ones occured the most, ignoring (for now) the instances where no name was provided:
